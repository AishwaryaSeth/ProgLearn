```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import pandas as pd

import matplotlib as mpl
import matplotlib.cm as cm

norm = mpl.colors.Normalize(vmin=0, vmax=8)
cmap = cm.rainbow

m = cm.ScalarMappable(norm=norm, cmap=cmap)
```

```python
n_shifts = 6

backward_dfs = []

for shift in range(1, n_shifts):
    df_of_split = pickle.load(open('shift_{}.p'.format(shift), 'rb'), encoding = 'latin1')
    backward_dfs.append(df_of_split)
    
    for task_num in range(10):
        df_of_split_of_task = df_of_split[df_of_split['task'] == task_num]
        if task_num < 9:
            org_accuracy = np.array(df_of_split_of_task['org_accuracy'])[0]
            if task_num == 1:
                plt.plot(range(task_num, 10), 
                         np.array(df_of_split_of_task['reverse_accuracy']),
                         c = m.to_rgba(shift),
                         label = "Data Fold: {}".format(shift)
                        )
            else:
                plt.plot(range(task_num, 10), 
                         np.array(df_of_split_of_task['reverse_accuracy']),
                         c = m.to_rgba(shift)
                        )
    plt.title("Accuracies Per Fold - CIFAR")
    plt.ylabel("Accuracy Per Task")
    plt.xlabel("Number of Tasks Seen")
    plt.tight_layout()
    plt.legend()
    
backward_df = pd.concat(backward_dfs)
```

```python
orig_accuracies_across_shifts = []
forward_accuracies_across_shifts = []

for shift in range(1, n_shifts):
    df_of_split = pickle.load(open('shift_{}.p'.format(shift), 'rb'), encoding = 'latin1')
    for task in range(10):
        orig_accuracy = df_of_split[(df_of_split['task'] == task) & (df_of_split['task_seen'] == task)]['org_accuracy']
        orig_accuracies_across_shifts.append(orig_accuracy)
    forward_accuracies = df_of_split[(df_of_split['task'] == 0)]['forward_accuracy']
    forward_accuracies_across_shifts.append(forward_accuracies)
        
mean_orig_accuracy = np.mean(orig_accuracies_across_shifts, axis = 0)
mean_forward_accuracies = np.mean(forward_accuracies_across_shifts, axis = 0)    
ftes = mean_forward_accuracies / mean_orig_accuracy 

plt.title("FTE Plot For Task 0 - CIFAR")
plt.ylabel("FTE")
plt.xlabel("Num Tasks Seen")
plt.plot(range(10), ftes)
plt.tight_layout()
plt.show()

```
