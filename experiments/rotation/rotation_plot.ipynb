```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
angles = np.arange(0,360,2)
mean_tes = []
for angle in angles:
    orig_error, transfer_error = pickle.load(open("rotation_results/angle_" + str(angle) + ".pickle", "rb"))
    mean_tes.append(orig_error / transfer_error)
```

```python
angles_l2n = np.arange(0,272,2)
mean_tes_l2n = []
for angle_l2n in angles_l2n:
    orig_error_l2n, transfer_error_l2n = pickle.load(open("rotation_results/angle_" + str(angle_l2n) + "_dnn.pickle", "rb"))
    mean_tes_l2n.append(orig_error_l2n / transfer_error_l2n)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.plot(angles,mean_tes, label = "L2F")
ax.plot(angles_l2n,mean_tes_l2n, label = "L2N")

#ax.set_yticks([1.0, 1.05, 1.1])
ax.tick_params(labelsize=20)
ax.set_xlabel('Angle of Rotation (Degrees)', fontsize=24)
ax.set_ylabel('Backward Transfer Efficiency', fontsize=24)
ax.set_title("Rotation Experiment", fontsize = 24)
ax.legend(fontsize = 24)
```
