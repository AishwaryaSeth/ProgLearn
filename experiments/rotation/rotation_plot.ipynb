```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
angles = np.arange(0,360,2)
mean_tes = []
for angle in angles:
    orig_error, transfer_error = pickle.load(open("rotation_results/angle_" + str(angle) + ".pickle", "rb"))
    mean_tes.append(orig_error / transfer_error)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.plot(angles,mean_tes)

#ax.set_yticks([1.0, 1.05, 1.1])
ax.tick_params(labelsize=20)
ax.set_xlabel('Angle of Rotation(Degree)', fontsize=24)
ax.set_ylabel('Transfer Efficiency', fontsize=24)
ax.set_title("L2F Rotation Experiment \n(ntrees = 16, reps = 20)", fontsize = 24)
```
